```python
import pandas as pd
```

```python
rawsettlementdata = pd.read_csv('/Users/julieagwu/Library/Mobile Documents/com~apple~CloudDocs/GB Power Market/Settlement Data/data/rawsettlement.csv')
```

```python
rawsettlementdata.head()
```

```python
rawsettlementdata.columns
```

```python
#only keep these columns except for StartTime, SystemBuyPrice, NetImbalanceVolume
#drop all columns from rawsettlementdata except StartTime, SystemBuyPrice, NetImbalanceVolume
rawsettlementdata.drop(rawsettlementdata.columns.difference(['StartTime', 'SystemBuyPrice', 'NetImbalanceVolume']), axis=1, inplace=True)
rawsettlementdata.head()
```

```python
#info() method is used to get a concise summary of the dataframe
rawsettlementdata.info()
```

```python
#changing the data type of StartTime to datetime
rawsettlementdata['StartTime'] = pd.to_datetime(rawsettlementdata['StartTime'])
```

```python
#split the StartTime column into two columns: Date and Time, then delete original StartTime column
rawsettlementdata['Date'] = rawsettlementdata['StartTime'].dt.date
rawsettlementdata['Time'] = rawsettlementdata['StartTime'].dt.time
rawsettlementdata.drop('StartTime', axis=1, inplace=True)
```

```python
rawsettlementdata.head()
```

```python
#In the Time column, the seconds are not needed, so we will remove them
#this code will remove the seconds from the Time column by converting the Time column to a string, then splitting the string at the colon, then joining the first two elements of the split string with a colon

rawsettlementdata['Time'] = rawsettlementdata['Time'].astype(str).str.split(':').str[0:2].str.join(':')

```

```python
rawsettlementdata.info()
```

```python
# make the time colume a hh:mm time datatype
rawsettlementdata['Time'] = pd.to_datetime(rawsettlementdata['Time'], format='%H:%M').dt.time
```
